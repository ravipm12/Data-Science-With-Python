# What is Regression  

Regression is a statistical method used in finance, investing, and other disciplines that attempts to determine the **strength and character of the relationship** between one dependent variable (usually denoted by Y) and a series of other variables (known as independent variables)  

**Two important terms in regression:**  
* Response, Outcome, or Dependent variable (y): The variable of interest that one wants to model or predict using one or more variables whose values are known.  
* Predictor, Explanatory, Feature or Independent variables(x) : It is assumed that the response depends on one or more predictors. These variables are independent and a model is formulated identifying the explicit relationship between the response and the predictor(s).  

**Types of Regression:**  
1. Simple linear regression : Summarize and study relationships between two continuous (quantitative) variables  
2. Multiple linear regression : Summrize and study relationships between one continuous response and multiple predictor variables  


Simple linear regression gets its adjective "simple," because it concerns the study of only one predictor variable. In contrast, multiple linear regression gets its adjective "multiple," because it concerns the study of two or more predictor variables

# What is Pairwise Correlation  
Before a regression model of alcohol on the predictors can be built, it is necessary to investigate whether there exists any dependence among the observed variables.  

**Formal definition of Correlation**: The pairwise correlation coefficient (denoted as ùëü) measures the degree of relatedness between two variables. 
The correlation coefficient (ùëü) developed by *Karl Pearson* is termed as Pearson product-moment correlation coefficient.   
The term ùëü is a measure of the strength and the direction of the linear relationship between **two continuous** variables  
  
Here, ùëã and ùëå are the pair of attributes measured on ùëõ units. From the i-th unit a pair of observations (ùë•ùëñ, ùë¶ùëñ) is obtained, and ùë•ÃÖ and ùë¶ÃÖ are the respective means. The value of ùëü ranges from ‚àí1 to +1, both values included.  

In general, there is no nice practical operational interpretation for r.You can only use r to make a statement about the strength of the linear relationship between x and y. 
In general:
* If r = -1, then there is a perfect negative linear relationship between x and y.If (x, y) pairs are plotted, the points would fall on a straight line with negative slope. This indicates that X and Y are perfectly linearly related but, if one variable increases, the other decreases  
* If r = 1, then there is a perfect positive linear relationship between x and y.If (x, y) pairs are plotted, the points would fall on a straight line with positive slope. This indicates that X and Y are perfectly linearly related and if one variable increases, the other also increases.  
* If r = 0, then there is no linear relationship between x and y.

All other values of r tell us that the relationship between x and y is not perfect.The closer r is to 0, the weaker the linear relationship.The closer r is to -1, the stronger the negative linear relationship.And, the closer r is to 1, the stronger the positive linear relationship.   

The plot of (x, y) continous variable pairs is known as a scatterplot.  
In the plots below , the 1st scatterplot corresponding to r = 0 clearly shows that there is no discernible pattern in the plot. The other two scatterplots exhibit moderate correlations of equal magnitude (0.6) but in opposite directions.

**PAIR PLOT WITH CORRELATION CO-EFFICIENTS**

```python
def corrdot(*args, **kwargs):
    corr_r = args[0].corr(args[1], 'pearson')
    corr_text = round(corr_r, 2)
    ax = plt.gca()
    font_size = abs(corr_r) * 80 + 5
    ax.annotate(corr_text, [.5, .5,],  xycoords="axes fraction",
                ha='center', va='center', fontsize=font_size)

def corrfunc(x, y, **kws):
    r, p = stats.pearsonr(x, y)
    p_stars = ''
    if p <= 0.05: ## Statistically significant @ alpha = 0.05
        p_stars = '*'
    if p <= 0.01: ## Statistically significant @ alpha = 0.01
        p_stars = '**'
    if p <= 0.001: ## Statistically significant @ alpha = 0.01
        p_stars = '***'
    ax = plt.gca()
    ax.annotate(p_stars, xy=(0.65, 0.6), xycoords=ax.transAxes,
                color='red', fontsize=70)

sns.set(style='white', font_scale=1.6)
g = sns.PairGrid(df2, aspect=1.5, diag_sharey=False, despine=False)
g.map_lower(sns.regplot, lowess=True, ci=False,
            line_kws={'color': 'red', 'lw': 1},
            scatter_kws={'color': 'blue', 's': 20})
g.map_diag(sns.distplot, color='black',
           kde_kws={'color': 'red', 'cut': 0.7, 'lw': 1},
           hist_kws={'histtype': 'bar', 'lw': 2,
                     'edgecolor': 'k', 'facecolor':'blue'})
g.map_diag(sns.rugplot, color='black')
g.map_upper(corrdot)
g.map_upper(corrfunc)
g.fig.subplots_adjust(wspace=0, hspace=0)

# Remove axis labels
for ax in g.axes.flatten():
    ax.set_ylabel('')
    ax.set_xlabel('')

# Add titles to the diagonal axes/subplots
for ax, col in zip(np.diag(g.axes), df2.columns):
    ax.set_title(col, y=0.82, fontsize=26)
```

## Important Note :
Correlation is only useful in determining linear relationship between two variables. Zero correlation may imply no linear dependence, but that does not preclude any other form of dependence (such as, polynomial) between the variables concerned.  

Further, correlation does not indicate any cause and effect relationship. Correlation simply quantifies how well two variables are related, if at all, and its direction (i.e. positive or negative).  

Another important observation about correlation is that, even when correlation is significantly different from zero (the * in correlation chart indicates precisely that), it may not have any practical significance.   

Typically, the rule of thumb for correlation values is as follows:  
1. r between ‚àí0.4 and +0.4 indicates absence of linear dependence  
2. r between ‚àí0.7 and ‚àí0.4 or r between +0.4 and +0.7 indicates moderate linear dependence, the sign indicating its direction  
3. r less than ‚àí0.7 or r greater than +0.7 indicates strong linear dependence

# Simple Linear Regression  

## Definition  

**The formal definition of Simple Linear Regression:** 
Let n pairs of observations (ùë•ùëñ, ùë¶ùëñ), i = 1, 2, ‚Ä¶, n, be available on two features, one of which is assumed to depend on the other. Typically Y denotes the dependent variable and X the independent variable. 
Let the quantity ùê∏(ùëå) denote the expected or mean value of ùëå.

**Simple linear regression relates the response Y to the single predictor variable X through a straight line. The mathematical formulation of the simple linear regression line is: ùê∏(ùëå)=ùõΩ0 +ùõΩ1 ùëã**  
* ùëå: is the value of the continuous response (or dependent) variable  
* ùõΩ0 and ùõΩ1: are intercept and slope coefficients, respectively, and known as the regression parameters.  
* ùëã: represents the independent (predictor) variable continuous in nature.  

It is assumed that the expected value of the response is a linear function of the predictor.  

When ùõΩ0 and ùõΩ1are known, a given value of the predictor will specify the expected value of the response.However, since Y is a random variable, not all values ùë¶ùëñ will be equal to ùê∏(ùëå).   

**Another form of SLR equation is ùëåùëñ=ùõΩ0 +ùõΩ1 ùëãùëñ + ùúÄùëñ, i = 1, 2, ‚Ä¶,n**   
ùúñ: represents the unobservable error term. Note that error here does not indicate any mistake, simply the difference between the expected and observed values of the response.The error terms provide crucial insight into the regression process.

## The method of Ordinary Least Square (OLS)  

The simple linear regression model involves unknown parameters ùõΩ0 and ùõΩ1, which need to be estimated from data. There are several different methods of estimating the parameters. The simplest and the most widely used method is known as the Ordinary Least Squares method (OLS).  
Given 1 pair of observations on ùëå and ùëã, the objective is to minimise the sum of squared errors and get appropriate estimates of ùõΩ0 and ùõΩ1.   
We want to minimise  
$$\displaystyle \sum _{i}{\hat {e}}_{i}^{2}=\sum _{i}({\hat {Y}}_{i}-({\hat {\beta }}_{0}+{\hat {\beta }}_{1}X_{1i})^{2}$$

### Fitting a Simple Regression Line

**REGRESSION WITH STATSMODEL**

**Step1 : Import package**

```python
import statsmodels.api as sm
```

**Step2 : Create the Feature and Dependent Variables**

```python
## Creating array of Feature X
X= df[['density']]

## Creating the array of Target y
y = df[['alcohol']]
```

**Step3 : Adding the Constant**  
By default, statsmodels fits a line passing through the origin, i.e. it doesn't fit an intercept. Hence, you need to use thecommand 'add_constant' so that it also fits an intercept.

```python
# add constant
X = sm.add_constant(X)
```

**Step4 : Fit the OLS Regression Model**  
We call the .fit() method to fit a model

```python
# fit the OLS Linear Regression model
Smodel = sm.OLS(y, X).fit()
```

**Step5 : Model Summary**

```python
# model summary
Smodel.summary()
```

2. const coefficient is your Y-intercept. It means that if Density coefficient is zero, then the expected output (i.e., the Y) would be equal to the const coefficient.  
3. Density coefficient represents the change in the output Y(% alcohol) due to a change of one unit in the Density (everything else held constant)  
4. std err reflects the level of accuracy of the coefficients. The lower it is, the higher is the level of accuracy  
5. P >|t| is your p-value. A p-value of less than 0.05 is considered to be statistically significant
6. Confidence Interval represents the range in which our coefficients are likely to fall (with a likelihood of 95%)

## The Coefficient of Determination ùêëùüê

**The coefficient of determination R2 is a summary measure that explains how well the sample regression line fits the data.**   
The rationale and computation of R2 is discussed below:
Let ùëåùëñÃÇ=ùõΩÃÇ0+ùõΩÃÇ1ùëãùëñ and ùëåÃÖ=1ùëõŒ£ùëåùëñùëõùëñ=1. ùëåùëñÃÇ is the estimated value of the response, for a given value of the predictor ùëãùëñ.

The difference between the observed and the estimated values of the response is called residual. Residual is the estimated value of the unobserved error component in the regression equation.ùúñÃÇùëñ = ùëåùëñ - ùëåÃÇùëñ  

Residuals are very important part of regression and they have many useful properties. 



1. SSR is the "regression sum of squares" and quantifies how far the estimated sloped regression line (y^),is from the horizontal "no relationship line," the sample mean or (ybar)

2. SSE is the "error sum of squares" and quantifies how much the data points(yi)vary around the estimated regression line (y^)  

3. SSTO/SST is the "total sum of squares" and quantifies how much the data points (yi, vary around their mean or (ybar) 


Note! that SSTO = SSR + SSE. The sums of squares appear to tell the story pretty well. They tell us that most of the variation in the response y (SSTO = 1827.6) is just due to random variation (SSE = 1708.5), not due to the regression of y on x (SSR = 119.1). You might notice that SSR divided by SSTO is 119.1/1827.6 or 0.065  

**The "coefficient of determination" or "r-squared value," denoted R2, is the regression sum of squares divided by the total sum of squares**  

**R2=(ùëÜùëÜùëÖ/ùëÜùëÜùëá) or R2=1‚àí(ùëÜùëÜùê∏/ùëÜùëÜùëá)**  

The numerical value of ùëÖ2 is equal to the square of the correlation coefficient between the response and the predictor only if there is a single predictor in the linear regression model, i.e. for simple linear regression only.  

1. ùëÖ2 measures the proportion of the total variation in ùëå that is explained by the regression model.  
2. ùëÖ2 ranges from 0‚â§ùëÖ2‚â§1. The higher the value of ùëÖ2, the more powerful is the predictor to predict the response.   
3. A regression model with high ùëÖ2 value indicates that the model fits the data well. In that case a high proportion of variance in the response is explained by the dependence of the response on the predictor

## R2 Cautions

Unfortunately, the coefficient of determination r2 and the correlation coefficient r have to be the most often misused and misunderstood measures in the field of statistics. To ensure that you don't fall victim to the most common mistakes, we review a set of seven different cautions here

**Caution 1**  
The coefficient of determination r2 and the correlation coefficient r quantify the strength of a linear relationship. It is possible that r2 = 0% and r = 0, suggesting there is no linear relation between x and y, and yet a perfect curved (or "curvilinear" relationship) exists.

**Caution 2**
A large r2 value should not be interpreted as meaning that the estimated regression line fits the data well. Another function might better describe the trend in the data.

**Caution 3**
The coefficient of determination r2 and the correlation coefficient r can both be greatly affected by just one data point (or a few data points) i.e. sensitive outliers and leverage points.

**Caution 4**
Correlation (or association) does not imply causation. Association between the predictor x and response y should not be interpreted as implying that x causes the changes in y. There are many possible reasons for why there is an association between x and y, including:

* The predictor x does indeed cause the changes in the response y.
* The causal relation may instead be reversed. That is, the response y may cause the changes in the predictor x.
* The predictor x is a contributing but not sole cause of changes in the response variable y.
* There may be a "lurking variable" that is the real cause of changes in y but also is associated with x, thus giving rise to the observed relationship between x and y.
* The association may be purely coincidental

**Caution 5**
A "statistically significant" r2 value does not imply that the slope Œ≤1 is meaningfully different from 0.

In general, the larger the data set, the easier it is to reject the null hypothesis and claim "statistical significance." If the data set is very large, it is even possible to reject the null hypothesis and claim that the slope Œ≤1 is not 0, even when it is not practically or meaningfully different from 0. It is possible to get a significant P-value when Œ≤1 is 0.13, a quantity that is likely not to be considered meaningfully different from 0. The key here is that "statistical significance does not imply practical significance."

**Caution 6** A large r2 value does not necessarily mean that a useful prediction of the response ynew, or estimation of the mean response ŒºY, can be made. It is still possible to get prediction intervals or confidence intervals that are too wide to be useful.

## Residual Plots    

Residuals are estimated errors and are defined as ùúñÃÇùëñ=ùëåùëñ‚àíùëåÃÇùëñ.
Residuals have many important properties and are employed to check various regression assumptions.
Recall that "Sum of Residuals = 0"
When conducting a residual analysis, a **residuals versus fits plot** is the most frequently created plot. It is a scatter plot of residuals on the y axis and fitted values (estimated responses) on the x axis.   
**The residulas versus fitted plot is used to detect non-linearity, unequal error variances, and outliers.**   

1. If the assumptions (Section7) of linear regression are satisfied, the residuals will form a band around the 0 line when plotted against the fitted values. The residuals should be randomly distributed around the horizontal line (that represents zero residual errors) i.e. there should not be a distinct trend in the distribution of points.  
2. If the scatterplot shows any systematic deviations from this band shape, they may indicate various departures from the assumptions. For example, if the residuals show any funnel-out or funnel-in shape, then one may conclude that the error variances are not all equal. If the scatterplot shows sinusoidal pattern, that may indicate serial dependence among the observations.  

**Charateristics of a well-behaved residual vs. fits plot and what they suggest about the appropriateness of the simple linear regression model:**

1. The residuals "bounce randomly" around the residual = 0 line. This suggests that the assumption that the relationship is linear is reasonable.
2. The residuals roughly form a "horizontal band" around the residual = 0 line. This suggests that the variances of the error terms are equal.
3. No one residual "stands out" from the basic random pattern of residuals. This suggests that there are no outliers

# Multiple Linear Regression (MLR)  
In the previous sections we have investigated how one single predictor variable may be used to predict a response. In reality, the response depends on multiple variables. For example, the volume of car sales of a company depends not only on the GDP but also on other factors like crude oil prices, confidence index of the population and many other considerations. In general, the response variable is found to depend on several predictors simultaneously.

Multiple linear regression is one of the simplest statistical tools that is used to analyse the relationship of the response with several predictors. Like simple linear regression, this procedure fits a linear function of the predictors to the response.

## Definition :  
The formal definition of Multiple Linear Regression: Multiple regression is a statistical technique used to analyse relationship between a single dependent variable and several predictors simultaneously. Assume n (multivariate) sample observations (ùë•1ùëñ, ùë•2ùëñ, ‚Ä¶ùë•ùëòùëñ, ùë¶ùëñ), i = 1, 2, ‚Ä¶, n, are available. Y denotes the dependent variable and ùëã1, ùëã2, ‚Ä¶,ùëãùëò the independent variables.  

The mathematical formulation of multiple linear regression line is: **ùê∏(ùëå)=ùõΩ0 +ùõΩ1 ùëã1+ùõΩ2ùëã2+‚ãØ+ùõΩùëòùëãùëò**  

* ùëå: is the value of the continuous response (or dependent) variable,
* ùõΩ0 : is the intercept
* ùëãùëó: represents the ùëóùë°‚Ñé independent (predictor) variable continuous in nature. j = 1, ‚Ä¶, k
* ùõΩùëó: represents the coefficient of the ùëóùë°‚Ñé independent (predictor) variable.  

It is assumed that the expected value of the response is a linear function of all the k predictors.  
When the regression coefficients ùõΩ0 and ùõΩùëó,ùëó = 1,‚Ä¶,ùëò are all known, or estimated, a given value of the predictor combination will specify the expected value of the response.  

Another form of MLR equation is **ùëåùëñ=ùõΩ0 +ùõΩ1 ùëã1ùëñ+ùõΩ2ùëã2ùëñ+‚ãØ+ùõΩùëòùëãùëòùëñ + ùúÄùëñ, i = 1, 2, ‚Ä¶,n**
ùúÄ: represents the unobservable error term.  

Formally the parameters ùõΩ1,ùõΩ2,‚Ä¶,ùõΩùëò are called **partial regression coefficients**, since the coefficient ùõΩùëò measures the change in the value of the response due to unit change in the value of ùëãùëò, keeping all other variables fixed.  
We will refer to the ùõΩ0 ,ùõΩ1,ùõΩ2,‚Ä¶,ùõΩùëò simply as regression coefficients.  

## The method of Ordinary Least Squares  
As in the case of simple linear regression , in multiple regression also the regression coefficients are estimated by minimizing the error sum of squares

### Dealing with Categorical Variables  
Before the problem of fitting a multiple linear regression model is taken up, the case of categorical predictor needs to be explicitly discussed.
So far we have tacitly assumed that the response as well as the predictors are all continuous.Let us discuss the situation when one or more predictors are categorical.   
A set of **dummy variables or indicator variables** is introduced corresponding to each categorical variable.   
**A discrete variable with ùëö categories is represented by a set of ùëö‚àí1 dummy variables.**

**Indicator coding / One-Hot-Coding : It is the most common format of dummy variable coding in which each category of the nominal variable is represented by either 1 or 0.**  

Let us now look at how Indicator Variables impact the regression equation:  
If ùëã1,‚Ä¶,ùëãùëò are continuous then we can write the regression model of ùëå on ùëã as: ùëå=ùõΩ0 +ùõΩ1ùëã1+‚ãØ+ùõΩùëòùëãùëò+ùúñ   
When we have a predictor variable ùëã that is is a categorical variable with ùëö levels, a small modification is required for the above regression model.  
In our case , Brand is a categorical predictor with three levels Grover Zampa, Seagram and Sula Vineyards. Hence ùëö‚àí1=3‚àí1=2 i.e. TWO DUMMY VARIABLES need to be introduced in the model.The model equation now is written as   
**ùëå=ùõΩ0+ùõΩ1ùëã1+‚ãØ+ùõΩùëòùëãùëò+ùõΩùëò+1ùê∑1+ùõΩùëò+2ùê∑2+ùúñ  where ùê∑1 and ùê∑2 are dummy variables taking values 1 or 0 as per the observation representation.**  

**Charateristics of Dummy Variables :**  
1. A dummy variable (aka indicator variable) is an variable artificially  created to represent an attribute with two or more distinct levels / categories.
2. Since Regression Analysis treats all independent variables in the analysis as numeric variables, we need dummy variables created for categorical variables.
3. Dummy variables assign numbers 0 or 1 to indicate membership to the category which is mutually exclusive and exhaustive.
4. The number of dummy variables necessary to represent a single attribute variable is equal to the number of levels minus 1.
5. One dummy variable cannot be a linear combination of another.

**get_dummies() method to create "indicator variables"**  
1. create dummy variables for multiple categories  
pd.get_dummies(DataFrame, columns=['Colname1'], drop_first=True , prefix='Colname')  
2. drop_first=True handles n - 1 i.e. drops the "original categorical column" and creates associated summy variables  
3. prefix adds the original column name to the dummified variable for ease of identification

For an explicitly declared categorical variable, the corresponding dummy variable assignment is automatically done. The default choice of the baseline is that level whose name comes first in lexicographic ordering. The default ordering can be changed.

## Multi-collinearity

In multiple regression, if one or more pairs of explanatory variables is highly correlated among themselves, then the phenomenon is known as multi-collinearity.Unfortunately, when it exists, it can wreak havoc on our analysis and thereby limit the research conclusions we can draw.  
**Effects of Multi-collinearity : It leads to inflated standard errors of the estimates of the regression coefficients, which in turn affects significance of the regression parameters. Often the signs of the regression coefficients may also change. As a result the regression model becomes non-reliable or lacks interpretability**  

**Types of multicollinearity:**  
1. *Structural multicollinearity:* This is a mathematical artifact caused by creating new predictors from other predictors ‚Äî such as creating the predictor from the predictor x e.g. Profit(x1) = Revenue(x2) - Cost(x3).  
2. *Data-based multicollinearity:* This is a result of a poorly designed experiment, reliance on purely observational data, or the inability to manipulate the system on which the data are collected.

In the case of structural multicollinearity, the multicollinearity is induced by what you have done. Data-based multicollinearity is the more troublesome of the two types of multicollinearity. Unfortunately, it is the type we encounter most often!

**What causes multi-collinearity:**  
1. Multicollinearity could exist because of the problems in the dataset at the time of creation. These problems could be because of poorly designed experiments, highly observational data, or the inability to manipulate the data: e.g. determining the electricity consumption of a household from the household income and the number of electrical appliances. Here, we know that the number of electrical appliances in a household will increase with household income. However, this cannot be removed from the dataset  
2. Multicollinearity could also occur when new variables are created which are dependent on other variables:e.g. creating a variable for BMI from the height and weight variables would include redundant information in the model  
3.  Including identical variables in the dataset: e.g.including variables for temperature in Fahrenheit and temperature in Celsius.  
4. Inaccurate use of dummy variables can also cause a multicollinearity problem. This is called the Dummy variable trap: e.g. in a dataset containing the status of marriage variable with two unique values: ‚Äòmarried‚Äô, ‚Äôsingle‚Äô. Creating dummy variables for both of them would include redundant information. We can make do with only one variable containing 0/1 for ‚Äòmarried‚Äô/‚Äôsingle‚Äô status.

**Detection of Multi-collinearity :** Some of the common methods used for detecting multicollinearity include :  
1. Correlation Matrix: We can start by computing the pairwise correlations among all the independent variables. The independent variables should not be highly (positive or negative) correlated. But this itself is not enough as the correlation matrix only detects high pairwise correlations. It is possible that even when no pairwise correlations are high, several moderately correlated pairs may give rise to multi-collinearity.  
2. The t-tests for each of the individual slopes are non-significant (P > 0.05), but the overall F-test for testing all of the slopes are simultaneously 0 is significant (P < 0.05).  
3. Variance Inflation factor: Variance inflation factors measure the inflation in the variances of the regression parameter estimates due to collinearities that exist among the predictors. It is a measure of how much the variance of the estimated regression coefficient Œ≤k is ‚Äúinflated‚Äù by the existence of correlation among the predictor variables in the model.  
* R^2 value is determined to find out how well an independent variable is described by the other independent variables. A high value of R^2 means that the variable is highly correlated with the other variables. This is captured by the VIF which is denoted by VIF = 1/(1-R^2).So, the closer the R^2 value to 1, the higher the value of VIF and the higher the multicollinearity with the particular independent variable  
**General Rule of thumb for VIF:**   
a. VIF  = 1 then there is no correlation among the kth predictor and the remaining predictor variables, and hence the variance of Œ≤ÃÇk is not inflated at all.   
b. VIF = 4 or nearly > 4, we say there is moderate VIF.    
c. VIF >= 10 , it shows signs of high multi-collinearity.  

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
# function to calculate VIF 
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
#For each X, calculate VIF and save in dataframe#
    return(vif)
```

```python
# calculete VIF
calc_vif(X)
```

## R2 and Adjusted R2  
Recall that in simple linear regression, ùëÖ2 is the square of the pairwise correlation coefficient between the single predictor X and the response Y. In MLR no such interpretation of ùëÖ2 holds.A general formulation valid for both simple and multiple regression is that ùëÖ2 equals the square of the correlation between the observed response ùëå and estimated/predicted response Yhat.Hence a high numerical value of ùëÖ2 gives an intuitive justification that the model works well since the observed response and the predicted responses are close.  
**One limitation of the coefficient of determination is that, its value increases as the number of independent variables in the model increases.** A good regression model should include only those predictors that are significantly different from 0. However, the numerical value of R2 is non-decreasing even if non-significant predictors are included in the model. Addition of non-significant predictors adversely affect the predictive quality of the model. Therefore, there is a need of another measure of model adequacy viz. Adjusted R2..  
**The adjusted R-squared adjusts for the number of predictors in the model. Importantly, its value increases only when the new predictor improves the model fit more than expected by chance alone. The adjusted R-squared value actually decreases when the new predictor doesn‚Äôt improve the model fit by a sufficient amount**
